```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from mps import BWMPS
```

```python
file = 'bridge.png'
img_full = mpimg.imread(file)[:, :, 0]
print("Image size:",img_full.shape)
plt.figure(figsize=[9,6])
plt.imshow(img_full, cmap='gray')
plt.axis('off');
```

```python
img_mps = BWMPS.from_matrix(img_full, norm = False, mode = "DCT")
```

```python
img_mps.compression_ratio()
```

```python
final_matrix = img_mps.mps_to_matrix()

plt.imshow(final_matrix, cmap = 'gray')
plt.axis('off');
```

```python
cutoff = 0.15

img_mps.compress(cutoff)
final_matrix = img_mps.mps_to_matrix()

plt.imshow(final_matrix, cmap = 'gray')
plt.axis('off');
```
