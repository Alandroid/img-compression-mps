```python
import json
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

from imgcompressionmps.utils.metrics import calc_mean_std
from imgcompressionmps.utils.filetools import find_project_root

project_root = find_project_root()
```

```python
file_prefix = project_root / "src/imgcompressionmps/evaluation/results"

with open(file_prefix / "Combined jsons/ds003799_0_25_100steps_to_0p1_DCT.json", "r") as f:
    MRI_DCT = json.load(f)

with open(file_prefix / "Combined jsons/ds003799_0_50_100steps_to_0p1_Std.json", "r") as f:
    MRI_Std = json.load(f)

with open(file_prefix / "Combined jsons/fMRI_0_20_100steps_to_0p1_DCT.json", "r") as f:
    fMRI_DCT = json.load(f)

with open(file_prefix / "Combined jsons/fMRI_0_45_100steps_to_0p1_Std.json", "r") as f:
    fMRI_Std = json.load(f)

with open(file_prefix / "MRI_slice_ds003799_0_to_50_100_steps_to_03_DCT.json", "r") as f:
    MRIslice_DCT = json.load(f)

with open(file_prefix / "MRI_slice_ds003799_0_to_50_100_steps_to_03_Std.json", "r") as f:
    MRIslice_Std = json.load(f)    
```

```python
key_y = "ssim_list"
MRI_DCT_mean, MRI_DCT_std, MRI_compfactor_DCT = calc_mean_std(MRI_DCT, 100, key_y= key_y)
MRI_Std_mean, MRI_Std_std, MRI_compfactor_Std = calc_mean_std(MRI_Std, 100, key_y= key_y)
fMRI_DCT_mean, fMRI_DCT_std, fMRI_compfactor_DCT = calc_mean_std(fMRI_DCT, 500, key_y= key_y)
fMRI_Std_mean, fMRI_Std_std, fMRI_compfactor_Std = calc_mean_std(fMRI_Std, 500, key_y= key_y)
MRIslice_DCT_mean, MRIslice_DCT_std, MRIslice_compfactor_DCT = calc_mean_std(MRIslice_DCT, 100, key_y= key_y)
MRIslice_Std_mean, MRIslice_Std_std, MRIslice_compfactor_Std = calc_mean_std(MRIslice_Std, 100, key_y= key_y)
```
